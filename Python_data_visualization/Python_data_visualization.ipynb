```python
#Kütüphaneleri import ediyoruz.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# İris datasetini import ediyoruz.

df = sns.load_dataset('iris')
df.head()
```

```python
# Lineplot

sns.lineplot(x=range (0, df.shape[0]), y='sepal_width', data=df)
```

```python
# Lineplot Kümeli şekilde

plt.figure(figsize=(7,7))
sns.lineplot(x=range (0, df.shape[0]), y='sepal_width', hue='species', data=df)
plt.show()
```

```python
# Histogram

sns.histplot(df['sepal_width'], color = 'orange')
plt.xlabel('Taç En')
plt.ylabel('Toplam çanak genişliği')
plt.title('Histogram')
plt.show()
```

```python
# Violin Plot

plt.figure(figsize=(7,7))
sns.violinplot(x='species', y='sepal_length', data=df)
plt.show()
```

```python
plt.figure(figsize=(7,7))
for col in df.columns[:-1]:
  sns.kdeplot(df[col], shade=True)
  plt.show()
```

```python
# Tips adlı bahşiş verisetini import edip scatter plot görselleştirmesi ile analiz edeceğiz.

import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
# Görseli biraz daha özelleştirmek istersek, mesela hangi gün daha çok bahşiş bırakılmış onu görmek için aşağıdaki kodu yazıyoruz.

sns.scatterplot(data=df, x = 'total_bill', y = 'tip', hue='day')
plt.legend(loc=2)
plt.show()
```
