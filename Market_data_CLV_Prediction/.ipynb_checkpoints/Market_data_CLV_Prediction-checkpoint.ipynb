```python
# İlk olarak kullanacağımız kütüphaneleri import ediyoruz.

import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from lifetimes import BetaGeoFitter
from lifetimes import GammaGammaFitter
from lifetimes.plotting import plot_period_transactions
from sklearn.preprocessing import MinMaxScaler
```

```python
# aykırı değerleri tespit etmesi için aşağıdaki fonksiyonu yazıyoruz.

def outliers(data,variable):
    Q1 = data[variable].quantile(0.01)
    Q3 = data[variable].quantile(0.99)
    interquantile = Q3 - Q1
    up_limit = Q3 + 1.5 * interquantile
    low_limit = Q3 - 1.5 * interquantile
    return up_limit, low_limit
```

```python
# Tespit edilen aykırı değerler aşağıdaki fonksiyon sayesinde olması gereken aralığa getirilir.

def outliers_replace(data,variable):
    up_limit,low_limit = outliers(data,variable)
    
    data.loc[(data[variable]>up_limit, variable)] = up_limit
    data.loc[(data[variable]<low_limit), variable] = low_limit
```

```python
# verisetini import ediyoruz.

df = pd.read_csv('datasets/kaggleSalesData.csv', encoding='Latin-1', sep=';')
df.head()
```

```python
# alt 5 satırı getiriyoruz.

df.tail()
```

```python
# Benzersiz user_id lerin sayısına bakıyoruz.

df['user_id'].nunique()
```

```python
# verisetinin içeriğini inceliyoruz.

df.info()
```

```python
# Yaşam boyu değeri tahmini için ihtiyacımız olan sütunları alıyoruz.

df = df[['DATE_','ORDERID','USERID','AMOUNT','PRICE','TOTALPRICE','USERGENDER']]
```

```python
# sütun adlarını anlaşılır şekle getiriyoruz.

df.columns = ['order_date','order_id','user_id','amount','price','total_price','gender']
```

```python
# verisetinin boyutuna bakıyoruz.

df.shape
```

```python
# Price sütununu numerik yapıyoruz.

df['price'] = df['price'].str.replace(',','.')
df['price'] = pd.to_numeric(df['price'])
```

```python
# Eksi değer çıkmaması adına amount ve price sütunlarını 0 dan büyük olarak alıyoruz.

df = df[df['amount']>0]
df = df[df['price']>0]
```

```python
# amount ve price değerlerini aykırı değerlerden arındırıyoruz.

outliers_replace(df,'amount')
outliers_replace(df,'price')
```

```python
# verisetini tekrar inceliyoruz.

df.info()
```

```python
# İşlem yapabilmek için, Date sütununu tarih tipine çeviriyoruz.

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
# Total Price sütununu drop yapıyoruz.

df.drop('total_price',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Drop ettiğimiz total price sütununu kendimiz hesaplayıp tekrar ekliyoruz.

df['total_price'] = df['amount'] * df['price']
df.head()
```

```python
# En son verilen siparişi buluyoruz.

df['order_date'].max()
```

```python
# frequency değeri için son yapılan alışverişi

today = dt.datetime(2023,8,18)
```

```python
# Recency için, müşterinin yaptığı ilk alışverişten son alışverişe kadar geçen zaman bulunur.
# T değeri için, analiz yapılan tarihten, müşterinin ilk yaptığı alışveriş tarihi çıkarılır.
# benzersiz order_id adedini buluyoruz.
# son olarak ta müşterilerin bize ödediği miktarın toplamını buluyoruz.

clv_df = df.groupby('user_id').agg({ 'order_date' : [lambda order_date : (order_date.max()-order_date.min()).days,
                                                   lambda order_date : (today-order_date.min()).days],
                                    'order_id' : lambda order_id : order_id.nunique(),
                                    'total_price' : lambda total_price : total_price.sum()
})
```

```python
clv_df.head()
```

```python
clv_df.columns = clv_df.columns.droplevel(0)
```

```python
clv_df.columns = ['recency','T','frequency','monetary']
```

```python
# Bu hesaplamalardan sonra çıkan tablo aşağıdaki gibi olacaktır.

clv_df
```

```python
# Çıkan sonuçları BG-NBD modeline uygun hale getirmeliyiz, bunun için çıkan sonuçlara aşağıdaki işlemleri yapıyoruz.

# İlk olarak satın alma başına monetary ortalamasını hesaplıyoruz.

clv_df['monetary'] = clv_df['monetary'] / clv_df['frequency']
```

```python
# Değerleri haftalık olarak çevirebilmek adına sonuçları 7 ye bölüyoruz.

clv_df['recency'] = clv_df['recency']/7
```

```python
# 1 den fazla alışveriş yapan müşterileri alıyoruz.

clv_df = clv_df[clv_df['frequency'] > 1]
```

```python
# Aynı şekilde T değerini de haftalık hesaplıyoruz.

clv_df['T'] = clv_df['T']/7
```

```python
# Yapılan hesaplamalardan sonra çıkan sonucu yazdırıyoruz.

clv_df.head()
```

```python
# describe ile istatiksel bir analiz yapıyoruz.

clv_df.describe().T
```

```python
# BG-NBD modeli ile müşterilerin şirkete 1 haftalık , 1 aylık ve 3 aylık getirebileceği sipariş adetlerini tahmin ediyoruz.
# ilk olarak bgf adlı değişkene betageofitter ı tanımlıyoruz.

bgf = BetaGeoFitter(penalizer_coef=0.001)
```

```python
clv_df
```

```python
# sonrasında modelimizi recency, frequency ve T değerleri üzerinden fit ediyoruz.
# görüldüğü gibi model, 19984 gözlem birimi üzerinden alpha beta ve r için parametreleri bulundu.

bgf.fit(clv_df['recency'], clv_df['frequency'], clv_df['T'])
```

```python
# Aşağıdaki kod satırında 1 hafta içerisinde sipariş verme olasılığı olan ilk 10 müşteriyi sıraladık.
# 672 id li müşterinin 1 hafta içerisinde alışveriş yapma olasılığı oldukça düşük olsa da %45 olarak görünüyor.

bgf.conditional_expected_number_of_purchases_up_to_time(1,
                                                        clv_df['frequency'],
                                                        clv_df['recency'],
                                                        clv_df['T']).sort_values(ascending=False).head(10)
```

```python
clv_df.shape
```

```python
# 1 ve 3 aylık tahmini alışveriş adetlerini hesaplıyoruz.

clv_df['expected_purc_1_month'] = bgf.predict(4,
                                            clv_df['frequency'],
                                            clv_df['recency'],
                                            clv_df['T'])
clv_df['expected_purc_3_month'] = bgf.predict(4*3,
                                            clv_df['frequency'],
                                            clv_df['recency'],
                                            clv_df['T'])
```

```python
clv_df
```

```python
clv_df.head()
```

```python
# Şirketin 1,3 ve 6 aylık tahmini toplam  satış adetleri tahminini yazdırıyoruz.

print ('Şirketin 1 ayda alacağı tahmini sipariş adedi',bgf.predict(4,
                                                                 clv_df['frequency'],
                                                                 clv_df['recency'],
                                                                 clv_df['T']).sum())
print ('Şirketin 3 ayda alacağı tahmini sipariş adedi',bgf.predict(4*3,
                                                                 clv_df['frequency'],
                                                                 clv_df['recency'],
                                                                 clv_df['T']).sum())
print ('Şirketin 6 ayda alacağı tahmini sipariş adedi',bgf.predict(4*6,
                                                                 clv_df['frequency'],
                                                                 clv_df['recency'],
                                                                 clv_df['T']).sum())
```

```python
# Bu analiz ve tahminlerden sonra gammagamma modelini hazırlıyoruz.
# Bu model ile müşterilerden gelecek ortalama tahmini karı bulacağız.

# ilk olarak modeli alıyoruz.

ggf = GammaGammaFitter(penalizer_coef=0.01)
```

```python
# sonrasında ise frequency ve monetary değerleri ile modeli fit ediyoruz.

ggf.fit(clv_df['frequency'],clv_df['monetary'])
```

```python
# sonrasında aşağıdaki kod satırı ile tahmini karı hesaplayarak büyükten küçüğe ilk 10 müşteriyi yazdırıyoruz.

ggf.conditional_expected_average_profit(clv_df['frequency'], 
                                        clv_df['monetary']).sort_values(ascending=False).head(10)
```

```python
# Çıkan sonucu expected average profit adında bir sütun oluşturup dataframe e ekliyoruz.

clv_df['expected_average_profit'] = ggf.conditional_expected_average_profit(clv_df['frequency'], 
                                        clv_df['monetary'])
```

```python
clv_df
```

```python
# Müşterinin yaşam boyu değerini tahmin etmek için aşağıdaki kod satırını yazıyoruz.

clv_df['clv'] = ggf.customer_lifetime_value(bgf,
                                  clv_df['frequency'],
                                  clv_df['recency'],
                                  clv_df['T'],
                                  clv_df['monetary'],
                                  time=12, # 12 aylık veri
                                  freq='W', # T nin frekans bilgisi
                                  discount_rate=0.01)
```

```python
clv_df.sort_values(by='clv', ascending=False).head(10)
```

```python
clv_df['segment'] = pd.qcut(clv_df['clv'], 4, labels=('D','C','B','A'))
```

```python
clv_df.head()
```

```python
clv_df.groupby('segment').agg({'count','mean','sum'})
```
